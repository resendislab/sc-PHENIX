```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/MAGIC_GSEA_CLUSTERS.csv')
```

```python
df= df.drop(['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6'],axis=1)
```

```python
df
```

```python
import seaborn as sns
```

```python
# pandas pivot
heatmap1_data = pd.pivot_table(df, values='NES',
                     index=['Gene Set'],
                     columns='Cluster ')
```

```python
heatmap1_data = heatmap1_data.fillna(value=0)
```

```python
heatmap1_data
```

```python
cg= sns.clustermap(heatmap1_data,dendrogram_ratio=0.01,cmap="BuGn",figsize=(2.9,15))
cg.ax_row_dendrogram.set_visible(False)
cg.ax_col_dendrogram.set_visible(False)
cg.cax.set_visible(True)
```
