```python
import pandas as pd
import numpy as np

def insertar_ceros_aleatorios(df, porcentaje=10, semilla=42):
    """
    Inserta ceros de manera aleatoria en un DataFrame basado en un porcentaje especificado y es reproducible por una semilla.

    Parámetros:
    df (pd.DataFrame): DataFrame original.
    porcentaje (int, opcional): Porcentaje de ceros a insertar. Por defecto es 10%.
    semilla (int, opcional): Semilla para el generador de números aleatorios. Por defecto es 42.

    Retorna:
    pd.DataFrame: DataFrame con ceros insertados aleatoriamente y reproducibilidad garantizada.
    """
    # Fijar la semilla del generador de números aleatorios para reproducibilidad
    np.random.seed(semilla)

    # Calcular el número total de elementos en el DataFrame
    total_elementos = df.size

    # Calcular el número de elementos a cambiar a cero
    num_ceros = int(total_elementos * (porcentaje / 100))

    # Generar pares de índices únicos
    indices = np.random.choice(total_elementos, num_ceros, replace=False)
    indices_fila = indices // df.shape[1]
    indices_columna = indices % df.shape[1]

    # Crear una copia del DataFrame para modificar
    df_modificado = df.copy()

    # Insertar ceros en los índices seleccionados
    df_modificado.values[indices_fila, indices_columna] = 0

    return df_modificado
```

```python
import pandas as pd
import numpy as np
import umap
from sc_PHENIX import run_pca, sc_PHENIX
from scipy.stats import spearmanr
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

def insertar_ceros_aleatorios(data, porcentaje, semilla):
    np.random.seed(semilla)
    data_dropout = data.copy()
    mask = np.random.rand(*data.shape) < (porcentaje / 100)
    data_dropout[mask] = 0
    return data_dropout

def run_analysis_dim_variation(real, porcentaje=80, seed=62, t=5, decay=5, n_pca=70, umap_dims=list(range(1, 101))):
    np.random.seed(seed)
    real = real.to_numpy() if isinstance(real, pd.DataFrame) else real

    results = pd.DataFrame(columns=['Method', 'UMAP_Dim', 'Pearson', 'Spearman', 'R2'])

    dropout = insertar_ceros_aleatorios(real, porcentaje=porcentaje, semilla=seed)

    for umap_dim in tqdm(umap_dims, desc='Processing UMAP Dimensions'):
        # Configurar y aplicar UMAP sin inicialización PCA
        umap_fit = umap.UMAP(n_components=umap_dim, n_neighbors=50, n_epochs=1000, random_state=1)
        umap_data = umap_fit.fit_transform(dropout)

        # Configurar y aplicar sc-PHENIX con UMAP sin inicialización PCA
        imputed_phenix_umap = sc_PHENIX(dropout, umap_data, t=t, knn=30, decay=decay)

        # Calcular métricas para sc-PHENIX con UMAP sin inicialización PCA
        rows = []
        for i in range(imputed_phenix_umap.shape[0]):
            real_sample = real[i, :]
            imputed_sample = imputed_phenix_umap[i, :]
            pearson = np.corrcoef(real_sample, imputed_sample)[0, 1]
            spearman = spearmanr(real_sample, imputed_sample).correlation
            r2 = pearson**2
            rows.append({'Method': 'sc-PHENIX UMAP', 'UMAP_Dim': umap_dim, 'Pearson': pearson, 'Spearman': spearman, 'R2': r2})
        results = pd.concat([results, pd.DataFrame(rows)], ignore_index=True)

        # Configurar y aplicar PCA
        pca_data = run_pca(dropout, n_components=n_pca, random_state=1)

        # Configurar y aplicar UMAP con inicialización PCA
        umap_fit_pca = umap.UMAP(n_components=umap_dim, n_neighbors=50, n_epochs=1000, random_state=1)
        pca_umap_data = umap_fit_pca.fit_transform(pca_data)

        # Configurar y aplicar sc-PHENIX con UMAP con inicialización PCA
        imputed_phenix_pca_umap = sc_PHENIX(dropout, pca_umap_data, t=t, knn=30, decay=decay)

        # Calcular métricas para sc-PHENIX con UMAP con inicialización PCA
        rows = []
        for i in range(imputed_phenix_pca_umap.shape[0]):
            real_sample = real[i, :]
            imputed_sample = imputed_phenix_pca_umap[i, :]
            pearson = np.corrcoef(real_sample, imputed_sample)[0, 1]
            spearman = spearmanr(real_sample, imputed_sample).correlation
            r2 = pearson**2
            rows.append({'Method': 'sc-PHENIX PCA-UMAP', 'UMAP_Dim': umap_dim, 'Pearson': pearson, 'Spearman': spearman, 'R2': r2})
        results = pd.concat([results, pd.DataFrame(rows)], ignore_index=True)

    # Guardar los resultados en un archivo
    filename = f"sc_phenix_results_seed{seed}_por{porcentaje}.xlsx"
    results.to_excel(filename, index=False)
    print(f"Results saved to {filename}")

    return results

# Ejemplo de uso de la función con datos reales
results = run_analysis_dim_variation(real=TrueEXpre, porcentaje=80, seed=62, t=5, decay=5, n_pca=30, umap_dims=list(range(1, 101)))
print(results.head())

# Visualización de los resultados
plt.figure(figsize=(18, 6))

# Subplot para Pearson correlation
plt.subplot(1, 3, 1)
sns.lineplot(data=results, x='UMAP_Dim', y='Pearson', hue='Method', marker='o')
plt.title('Pearson Correlation vs UMAP Dimensions')
plt.xlabel('UMAP Dimensions')
plt.ylabel('Pearson Correlation')

# Subplot para Spearman correlation
plt.subplot(1, 3, 2)
sns.lineplot(data=results, x='UMAP_Dim', y='Spearman', hue='Method', marker='o')
plt.title('Spearman Correlation vs UMAP Dimensions')
plt.xlabel('UMAP Dimensions')
plt.ylabel('Spearman Correlation')

# Subplot para R2 Score
plt.subplot(1, 3, 3)
sns.lineplot(data=results, x='UMAP_Dim', y='R2', hue='Method', marker='o')
plt.title('R2 Score vs UMAP Dimensions')
plt.xlabel('UMAP Dimensions')
plt.ylabel('R2 Score')

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import numpy as np
import umap
from sc_PHENIX import run_pca, sc_PHENIX
from scipy.stats import spearmanr
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE

def insertar_ceros_aleatorios(data, porcentaje, semilla):
    np.random.seed(semilla)
    data_dropout = data.copy()
    mask = np.random.rand(*data.shape) < (porcentaje / 100)
    data_dropout[mask] = 0
    return data_dropout

def run_analysis_dim_variation_tsne(real, porcentaje=80, seed=62, t=5, decay=5, n_pca=70, tsne_dims=list(range(1, 101))):
    np.random.seed(seed)
    real = real.to_numpy() if isinstance(real, pd.DataFrame) else real

    results = pd.DataFrame(columns=['Method', 'tSNE_Dim', 'Pearson', 'Spearman', 'R2'])

    dropout = insertar_ceros_aleatorios(real, porcentaje=porcentaje, semilla=seed)

    for tsne_dim in tqdm(tsne_dims, desc='Processing tSNE Dimensions'):
        # Configurar y aplicar t-SNE sin inicialización PCA
        tsne_fit = TSNE(n_components=tsne_dim, random_state=1)
        tsne_data = tsne_fit.fit_transform(dropout)

        # Configurar y aplicar sc-PHENIX con t-SNE sin inicialización PCA
        imputed_phenix_tsne = sc_PHENIX(dropout, tsne_data, t=t, knn=30, decay=decay)

        # Calcular métricas para sc-PHENIX con t-SNE sin inicialización PCA
        rows = []
        for i in range(imputed_phenix_tsne.shape[0]):
            real_sample = real[i, :]
            imputed_sample = imputed_phenix_tsne[i, :]
            pearson = np.corrcoef(real_sample, imputed_sample)[0, 1]
            spearman = spearmanr(real_sample, imputed_sample).correlation
            r2 = pearson**2
            rows.append({'Method': 'sc-PHENIX tSNE', 'tSNE_Dim': tsne_dim, 'Pearson': pearson, 'Spearman': spearman, 'R2': r2})
        results = pd.concat([results, pd.DataFrame(rows)], ignore_index=True)

        # Configurar y aplicar PCA
        pca_data = run_pca(dropout, n_components=n_pca, random_state=1)

        # Configurar y aplicar t-SNE con inicialización PCA
        tsne_fit_pca = TSNE(n_components=tsne_dim, random_state=1,)
        pca_tsne_data = tsne_fit_pca.fit_transform(pca_data)

        # Configurar y aplicar sc-PHENIX con t-SNE con inicialización PCA
        imputed_phenix_pca_tsne = sc_PHENIX(dropout, pca_tsne_data, t=t, knn=30, decay=decay)

        # Calcular métricas para sc-PHENIX con t-SNE con inicialización PCA
        rows = []
        for i in range(imputed_phenix_pca_tsne.shape[0]):
            real_sample = real[i, :]
            imputed_sample = imputed_phenix_pca_tsne[i, :]
            pearson = np.corrcoef(real_sample, imputed_sample)[0, 1]
            spearman = spearmanr(real_sample, imputed_sample).correlation
            r2 = pearson**2
            rows.append({'Method': 'sc-PHENIX PCA-tSNE', 'tSNE_Dim': tsne_dim, 'Pearson': pearson, 'Spearman': spearman, 'R2': r2})
        results = pd.concat([results, pd.DataFrame(rows)], ignore_index=True)

    # Guardar los resultados en un archivo
    filename = f"sc_phenix_tsne_results_seed{seed}_por{porcentaje}.xlsx"
    results.to_excel(filename, index=False)
    print(f"Results saved to {filename}")

    return results

# Ejemplo de uso de la función con datos reales
results = run_analysis_dim_variation_tsne(real=TrueEXpre, porcentaje=80, seed=62, t=5, decay=5, n_pca=30, tsne_dims=list(range(1, 4)))
print(results.head())

# Visualización de los resultados
plt.figure(figsize=(18, 6))

# Subplot para Pearson correlation
plt.subplot(1, 3, 1)
sns.lineplot(data=results, x='tSNE_Dim', y='Pearson', hue='Method', marker='o')
plt.title('Pearson Correlation vs tSNE Dimensions')
plt.xlabel('tSNE Dimensions')
plt.ylabel('Pearson Correlation')


# Subplot para Spearman correlation
plt.subplot(1, 3, 2)
sns.lineplot(data=results, x='tSNE_Dim', y='Spearman', hue='Method', marker='o')
plt.title('Spearman Correlation vs tSNE Dimensions')
plt.xlabel('tSNE Dimensions')
plt.ylabel('Spearman Correlation')



# Subplot para R2 Score
plt.subplot(1, 3, 3)
sns.lineplot(data=results, x='tSNE_Dim', y='R2', hue='Method', marker='o')
plt.title('R2 Score vs tSNE Dimensions')
plt.xlabel('tSNE Dimensions')
plt.ylabel('R2 Score')


plt.tight_layout()
plt.show()
```
